```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import os
from os import listdir
```

```python
import os
os.getcwd()
collection = 'C:/Users/rammo/DIP/Assignment1/images1'
for i, filename in enumerate(os.listdir(collection)):
    os.rename('C:/Users/rammo/DIP/Assignment1/images1/' + filename, 'C:/Users/rammo/DIP/Assignment1/images1/r' + str(i) + ".jpg")
```

```python
path = r'C:\Users\rammo\DIP\Assignment1\images1'

img_org = []
for images in os.listdir(path):
    if( images.endswith(".jpg") or images.endswith(".jpeg")):
        img_org.append(images)
```

```python
img = img_org[:99]
img_ref = []
img_ref.append(img_org[99])

i2 = cv2.imread("images1/"+img_ref[0])
i2.shape
```

```python
rmse = []
for i in range(0, len(img)):
    i1 = cv2.imread("images1/"+img[i])
    i1 = cv2.resize(i1, (1200, 1600))
    i2 = cv2.imread("images1/"+img_ref[0])
    
    # print("Shape of 1 "+str(i1.shape)+"2: "+str(i2.shape))
    ms = np.mean(cv2.absdiff(i1,i2)**2)
    rmse.append(ms**0.5)
    
#     cv2.imshow('i1', i1)
#     cv2.imshow('i2', i2)
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()
# print(rmse)
```

```python
plt.figure(dpi=100, figsize=(7,5))
plt.bar(range(0,len(img)), rmse, color='red')
# plt.plot(range(0,len(img)), rmse, color='red')
plt.xlabel('Image')
plt.ylabel('RMSE')
plt.title('RMSE vs Image')
plt.show()
```

```python
plt.figure(dpi=100, figsize=(7,5))
plt.scatter(range(0,len(img)), rmse, color='red')
plt.plot(range(0,len(img)), rmse, color='red')
plt.xlabel('Image')
plt.ylabel('RMSE')
plt.title('RMSE vs Image')
plt.show()
```
